```python
import seaborn as sns

# Creating a correlation matrix based on the 'charges' column and other numerical columns
corr_matrix = cleand_df.corr()

# Plotting the heatmap
plt.figure(figsize=(10, 8))  # Adjust the figure size if needed
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.4)
plt.title('Correlation Heatmap')
plt.show()
```

#### `age`-- 'Age' factor have a strong impact on charges. Older individuals tend to have higher charges compared to younger ones.  
#### `smoker`-- 'Smoker' factor have a strong impact on charges.The difference in charges between smokers and non-smokers is Smokers have significantly higher charges compared to non-smokers.    
#### `sex`-- There is no relationship beteen 'sex' and 'charges'. Charges doesn't vary between males and females and are similar for both.  
#### `bmi`-- Their is no relationship between 'bmi' and 'charges'.  
#### `regions`-- There are no significant regional(northeast, southeast, southwest, northwest)  variations The average charges or the distribution of charges across different regions is almost similar. Compare to others,southeast region consumes high range of charges.   
#### `children` -- cildren feature having no Impact on charges.
